```python
top_student_df = df[df['Exam_Score'] == df['Exam_Score'].max()]
top_student_df
```

--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Adding a improvement status to show the progress of the student

```python

df['Diffrence In Exam_Scores'] = df['Exam_Score'] - df['Previous_Scores']
df.loc[df['Diffrence In Exam_Scores'] == 0, 'Improvment_Status'] = "No Improvement"
df.loc[df['Diffrence In Exam_Scores'] > 0, 'Improvment_Status'] = "Improvement"
df.loc[df['Diffrence In Exam_Scores'] < 0, 'Improvment_Status'] = "Improvement Decline"

df
```

Showing how many students have improved from the previous exam

```python
df['Improvment_Status'].value_counts()
```

Graphs:

```python
graph_1 = df.groupby('Peer_Influence')['Exam_Score'].mean()

graph_1.plot(kind='bar')
plt.title('Average Exam Score by Peer Influence')
plt.xlabel('Peer Influence')
plt.ylabel('Average Exam Score')
plt.savefig("Peer Influence vs Average Exam Score.png")

plt.show()
```

```python
graph_2 = df.groupby('Sleep_Hours')['Exam_Score'].mean()
graph_2.plot(kind='bar')

plt.title("Average Exam Score by Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Average Exam Score")
plt.savefig("Sleep Hours vs Exam Score.png")
```

```python
graph_3 = df['Improvment_Status'].value_counts()
graph_3.plot(kind='pie', autopct='%1.0f%%')
plt.savefig("Improvment_Status.png")

plt.show()
```
